```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
def k(x,y):
    return np.exp(-(x-y)**2)


X = np.linspace(0,5,10)
K = np.zeros([X.shape[0],X.shape[0]])
for i,x in enumerate(X):
    for j,y in enumerate(X):
        K[i,j] = 0.9*k(x,y)
```

```python
plt.imshow(K)
plt.colorbar()
plt.xticks([0,2,4,6,8],['10am','11am','noon','1pm','2pm'])
plt.yticks([0,2,4,6,8],['10am','11am','noon','1pm','2pm'])
plt.savefig('timedemo1.png')
```

```python
X1 = np.linspace(0,5,5)
X2 = np.linspace(0,5,10)
X = np.r_[X1,X2]
K = np.zeros([X.shape[0],X.shape[0]])
for i,x in enumerate(X):
    for j,y in enumerate(X):
        K[i,j] = k(x,y)
K[0:5,5:]*=0.5
K[5:,0:5]*=0.5
```

```python
plt.imshow(K)
plt.yticks([0,1,2,3,4,5,7,9,11,13],['10am','11am','noon','1pm','2pm','10am','11am','noon','1pm','2pm'])
plt.xticks([0,2,5,7,9,11,13],['10am','noon','10am','11am','noon','1pm','2pm'])
#plt.savefig('twosensors.pdf')
```
